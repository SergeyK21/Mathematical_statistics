```python
import math

from math import factorial


def c(n, k):
    return factorial(n) // (factorial(k) * factorial(n - k))


def arithmetic_mean(a: list):
    return sum(a) / len(a)


def mean_square_deviation(a: list):
    m = sum(a) / len(a)
    return math.sqrt(sum(list(map(lambda x: (x - m) ** 2, a))) / len(a))


def biased_variance_estimates(a: list):
    m = sum(a) / len(a)
    return sum(map(lambda x: (x - m) ** 2, a)) / len(a)


def unbiased_variance_estimates(a: list):
    m = sum(a) / len(a)
    return sum(map(lambda x: (x - m) ** 2, a)) / (len(a) - 1)
```

1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. 
Посчитать (желательно без использования статистических методов наподобие std, var, mean) 
среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
a = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

print(f'среднее арифметическое - {arithmetic_mean(a): .4f}')
print(f'среднее квадратичное отклонение - {mean_square_deviation(a): .4f}')
print(f'смещенная оценка дисперсии - {biased_variance_estimates(a): .4f}')
print(f'несмещенная оценка дисперсии - {unbiased_variance_estimates(a): .4f}')
```

3. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, 
сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. 
Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. 
Студент сдал первую сессию. Какова вероятность, что он учится: 
a). на факультете A б). на факультете B в). на факультете C?

```python
a = .8
b = .7
c = .9

# найдем общее кол-во студентов на 3-х факультетах x + x + 2*x = 1
# x = 1/4
# тогда вкроятность того что это студент факультета А равна = x = 1/4;
# вкроятность того что это студент факультета B равна = x = 1/4;
# вкроятность того что это студент факультета C равна = x = 1/2;

a_x = 1 / 4
b_x = 1 / 4
c_x = 1 / 2

PA = a_x * a + b_x * b + c_x * c

print(f'Вероятность, что студент сдал сессию и учится на факультете А - '
      f'{a_x * a / PA: .4f}')

print(f'Вероятность, что студент сдал сессию и учится на факультете B - '
      f'{b_x * b / PA: .4f}')

print(f'Вероятность, что студент сдал сессию и учится на факультете А - '
      f'{c_x * c / PA: .4f}')
```
