```python
import math


def get_Z(x, m, y):
    return (x - m) / y
```

1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

```python
print(f'среднее значение - {(200 + 800) / 2}')
print(f'Дисперсия - {(800 - 200) / 12}')
```

2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

```python
# D(x) = (b - a) / 12 => 0.2 = (b - 0.5)^2 / 12 => b = sqrt(0.2 * 12) + 0.5

b = math.sqrt(.2 * 12) + .5
print(f'Правая граница равна = {b}')

# M(x) = (0.5 + b) / 2
print((0.5 + b) / 2)
```

Найдите:

4. Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.

```python
m = 174
y = 8
# a) P(A) = 1 - P(x <= 182) => P(A) = 1 - P(z_a)

z_a = get_Z(182, m, y)
print(f'a) Z = (182 - 174) / 8 = {get_Z(182, m, y)}')

# z_a = 1 => P(z_a) = 0.3413

print(f'a) Вероятность = {1 - .3413: .4f}')

print('_________________________________________________')

# б) P(A) = 1 - P(x <= 190) => P(A) = 1 - P(z_б)

z_a = get_Z(182, m, y)

print(f'б) Z = (190 - 174) / 8 = {get_Z(190, m, y)}')

# z_a = 1 => P(z_б) = 0.4773

print(f'б) Вероятность = {1 - .4773: .4f}')

print('_________________________________________________')

print(f'в) Z1 = (166 - 174) / 8 = {get_Z(166, m, y)}')

print(f'в) Z2 = (190 - 174) / 8 = {get_Z(190, m, y)}')

print(f'в) P(Z1) = {.1587}')

print(f'в) P(Z2) = {.4773}')

print(f'в) P(A) = P(Z2) - P(Z1) = {.4773 - .1587}')

print('_________________________________________________')

print(f'г) Z1 = (166 - 174) / 8 = {get_Z(166, m, y)}')

print(f'г) Z2 = (182 - 174) / 8 = {get_Z(182, m, y)}')

print(f'г) P(Z1) = {.1587}')

print(f'г) P(Z2) = {.3413}')

print(f'г) P(A) = P(Z2) - P(Z1) = {.3413 - .1587: .4f}')

print('_________________________________________________')
print(f'д) Z1 = (158 - 174) / 8 = {get_Z(158, m, y)}')

print(f'д) Z2 = (190 - 174) / 8 = {get_Z(190, m, y)}')

print(f'д) P(Z1) = {.0228}')

print(f'д) P(Z2) = {.4773}')

print(f'д) P(A) = P(Z2) - P(Z1) = {.4773 - .0228: .4f}')

print('_________________________________________________')

print(f'е) Z1 = (150 - 174) / 8 = {get_Z(150, m, y)}')

print(f'е) Z2 = (190 - 174) / 8 = {get_Z(190, m, y)}')

print(f'е) P(Z1) = {.0014}')

print(f'е) P(Z2) = {.4773}')

print(f'е) P(A) = P(Z1) + 1 - P(Z2) = {.0014 + 1 - .4773: .4f}')

print('_________________________________________________')

print(f'ё) Z1 = (150 - 174) / 8 = {get_Z(150, m, y)}')

print(f'ё) Z2 = (198 - 174) / 8 = {get_Z(198, m, y)}')

print(f'ё) P(Z1) = {.0014}')

print(f'ё) P(Z2) = {.4987}')

print(f'ё) P(A) = P(Z1) + 1 - P(Z2) = {.0014 + 1 - .4987: .4f}')

print('_________________________________________________')

print(f'ж) Z = (166 - 174) / 8 = {get_Z(166, m, y)}')

print(f'ж) P(A) = P(Z) = {.1587}')
```

5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
print(f'кол-во сигм = {(190 - 178) / math.sqrt(25): .0f}')
```
